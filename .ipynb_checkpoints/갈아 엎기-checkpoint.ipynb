# 시간의 경우에는 24 는 datetime format 으로 인식하지 못하기 때문에 -1 을 진행한다. (의미는 비슷하기에 그대로 진행해도 된다)

흠.. 어떻게 맨 위에있는 것을 없애고 새롭게 저장하지... 

```python
df['dateInt']= df['연'].astype(str) + df['월'].astype(str).str.zfill(2)+df['일'].astype(str).str.zfill(2)+df['시간'].astype(str).str.zfill(2)
df['dateInt']
```

아 그냥 했더니 되는거 같네. 

```python
df['dateInt'].head()
```

```python
pd.to_datetime('2006-07-10 14:00', format="%Y-%m-%d %H:")
```

```python
df['Date'] = pd.to_datetime(df['dateInt'], format="%Y%m%d%H")
df['Date']
```

```python
df.head()
```

```python
df.insert(4, 'date', value = df['Date'])

df.head()
```

```python
df =df.drop(['연','월','일','시간','dateInt','Date'], axis = 1)
df.head()
```
