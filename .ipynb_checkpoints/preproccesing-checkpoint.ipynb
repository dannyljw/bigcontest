```python
df = pd.read_excel('datetime,avgadd.xlsx')
df.head()
```

```python
norm_inflow = (df['유입량']-df['유입량'].min())/(df['유입량'].max()-df['유입량'].min())
df['유입량']=norm_inflow
df.head()
```
