흠.. 어떻게 맨 위에있는 것을 없애고 새롭게 저장하지... 

```python
df['dateInt'] = df['연'].astype(str) + df['월'].astype(str).str.zfill(2)+df['일'].astype(str).str.zfill(0)
df['dateInt'] = df['dateInt'].drop([0])
df['dateInt']
```

아 그냥 했더니 되는거 같네. 

```python
df['Date'] = pd.to_datetime(df['dateInt'].drop([0]), format='%Y%m%d')
df['Date']
```

```python
df.insert(4, 'date', value = df['Date'])
```
